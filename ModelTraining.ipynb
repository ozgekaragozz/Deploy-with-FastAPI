```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder

label_encoder = LabelEncoder()

data['Sex'] = label_encoder.fit_transform(data['Sex'])  
data['ExerciseAngina'] = label_encoder.fit_transform(data['ExerciseAngina']) 

data = pd.get_dummies(data, columns=['ChestPainType', 'RestingECG'], drop_first=True)

ordinal_encoder = OrdinalEncoder(categories=[['Up', 'Flat', 'Down']])
data['ST_Slope'] = ordinal_encoder.fit_transform(data[['ST_Slope']])
```

```python
data.shape
```

```python
X = data.drop(['HeartDisease'],axis=1)
y = data['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

numeric_cols.remove('HeartDisease')

X_train[numeric_cols] = sc.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = sc.transform(X_test[numeric_cols])
```

```python
!pip install scikit-optimize
```

```python
from xgboost import XGBClassifier
from skopt import BayesSearchCV

classifier = XGBClassifier(scale_pos_weight=(len(y_train) - sum(y_train)) / sum(y_train))

param_dist_xgb = {
    'max_depth': (3, 10),
    'learning_rate': (0.01, 1.0, 'log-uniform'),
    'n_estimators': (50, 500),
    'gamma': (0, 1.0, 'uniform'),
    'subsample': (0.5, 1.0, 'uniform'),
    'colsample_bytree': (0.5, 1.0, 'uniform')
}

bayes_search_xgb = BayesSearchCV(
    estimator=classifier,
    search_spaces=param_dist_xgb,
    n_iter=30,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    random_state=42
)

bayes_search_xgb.fit(X_train, y_train)
best_params_bayes_xgb = bayes_search_xgb.best_params_
print(f"Bayesian Search ile bulunan en iyi hiperparametreler (XGBoost): {best_params_bayes_xgb}")
```

```python
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score

classifier = XGBClassifier(**best_params_bayes_xgb)

accuracy_scorer = make_scorer(accuracy_score)
precision_scorer = make_scorer(precision_score, average='weighted')
recall_scorer = make_scorer(recall_score, average='weighted')
f1_scorer = make_scorer(f1_score, average='weighted')

cross_val_results_train = {'Metric': [], 'Score': []}

accuracy_scores = cross_val_score(classifier, X_train, y_train, cv=5, scoring=accuracy_scorer)
precision_scores = cross_val_score(classifier, X_train, y_train, cv=5, scoring=precision_scorer)
recall_scores = cross_val_score(classifier, X_train, y_train, cv=5, scoring=recall_scorer)
f1_scores = cross_val_score(classifier, X_train, y_train, cv=5, scoring=f1_scorer)

cross_val_results_train['Metric'].extend(['Accuracy', 'Precision', 'Recall', 'F1 Score'])
cross_val_results_train['Score'].extend([
    f"{np.mean(accuracy_scores):.4f} ± {np.std(accuracy_scores):.4f}",
    f"{np.mean(precision_scores):.4f} ± {np.std(precision_scores):.4f}",
    f"{np.mean(recall_scores):.4f} ± {np.std(recall_scores):.4f}",
    f"{np.mean(f1_scores):.4f} ± {np.std(f1_scores):.4f}"
])

cross_val_results_train_df = pd.DataFrame(cross_val_results_train)
print(cross_val_results_train_df)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc, RocCurveDisplay

classifier = XGBClassifier(**best_params_bayes_xgb)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(ax=axes[0], cmap='Blues')
axes[0].set_title('XGBoost - Confusion Matrix')

if len(set(y_test)) == 2:
    fpr, tpr, _ = roc_curve(y_test, classifier.predict_proba(X_test)[:, 1])
    roc_auc = auc(fpr, tpr)
    RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name="XGBoost").plot(ax=axes[1])
    axes[1].set_title(f'XGBoost - ROC Curve (AUC = {roc_auc:.2f})')
else:
    axes[1].axis('off')  

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score

threshold = 0.85
high_corr_features = [
    column for column in corr_matrix.columns
    if any(corr_matrix[column] > threshold) and column != corr_matrix[column].idxmax()
]

X_train_filtered = X_train.drop(columns=high_corr_features)
X_test_filtered = X_test.drop(columns=high_corr_features)

remaining_features = X_train_filtered.columns.tolist()

print(f"Kalan Özellik Sayısı: {X_train_filtered.shape[1]}")
print("Çıkarılan Yüksek Korelasyonlu Özellikler:", high_corr_features)
print("Kalan Özellikler:", remaining_features)

accuracy_scorer = make_scorer(accuracy_score)
precision_scorer = make_scorer(precision_score, average='weighted')
recall_scorer = make_scorer(recall_score, average='weighted')
f1_scorer = make_scorer(f1_score, average='weighted')

xgb_model = XGBClassifier(**best_params_bayes_xgb)

cross_val_results_train = {'Metric': [], 'Score': []}

accuracy_scores = cross_val_score(xgb_model, X_train_filtered, y_train, cv=5, scoring=accuracy_scorer)
precision_scores = cross_val_score(xgb_model, X_train_filtered, y_train, cv=5, scoring=precision_scorer)
recall_scores = cross_val_score(xgb_model, X_train_filtered, y_train, cv=5, scoring=recall_scorer)
f1_scores = cross_val_score(xgb_model, X_train_filtered, y_train, cv=5, scoring=f1_scorer)

cross_val_results_train['Metric'].extend(['Accuracy', 'Precision', 'Recall', 'F1 Score'])
cross_val_results_train['Score'].extend([
    f"{np.mean(accuracy_scores):.4f} ± {np.std(accuracy_scores):.4f}",
    f"{np.mean(precision_scores):.4f} ± {np.std(precision_scores):.4f}",
    f"{np.mean(recall_scores):.4f} ± {np.std(recall_scores):.4f}",
    f"{np.mean(f1_scores):.4f} ± {np.std(f1_scores):.4f}"
])

cross_val_results_train_df = pd.DataFrame(cross_val_results_train)
print(cross_val_results_train_df)
```

```python
xgb_model = XGBClassifier(**best_params_bayes_xgb)
xgb_model.fit(X_train_filtered, y_train)

xgb_importance = xgb_model.get_booster().get_score(importance_type='weight')

xgb_importance_df = pd.DataFrame(list(xgb_importance.items()), columns=['Feature', 'Importance'])
xgb_importance_df = xgb_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(8, 6))
plt.barh(xgb_importance_df['Feature'], xgb_importance_df['Importance'], color='lightcoral')
plt.title('XGBoost Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```
